##Definindo o sistema

```
#Definindo os sistemas
num1=[15]
den1=[1, 10, 15]
sys1= TransferFunction(num1,den1)

num2=[25]
den2=[1, 10, 25]
sys2= TransferFunction(num2,den2)

num3=[60]
den3=[1, 10, 60]
sys3= TransferFunction(num3,den3)
```

##Plotando os gráficos

```
num_step = [1]
den_step = [1,0]
step = signal.TransferFunction(num_step,den_step)

t = np.linspace(0,4,100)

t,y1 = signal.step(signal.TransferFunction(num1,den1),0,t)
t,y2 = signal.step(signal.TransferFunction(num2,den2),0,t)
t,y3 = signal.step(signal.TransferFunction(num3,den3),0,t)

t,y_stp = signal.impulse(step,0,t) 

plt.plot(t,y_stp, "--y", label='degrau')
plt.plot(t, y1,  "-r", label='sistma 1')
plt.plot(t, y2,  "-k", label='sistema 2')
plt.plot(t, y3,  "-c", label='sistema 3')

plt. legend(loc = 'best')
plt.grid(axis='x', color='0.95')

plt.show()
```

##Definindo os sistemas

```
#Definindo os sistemas
num1=[4.8]
den1=[1, 2, 4.8]
sys1= control.TransferFunction(num1,den1)

num2=[2.8]
den2=[1, 2, 2.8]
sys2= control.TransferFunction(num2,den2)

num3=[2]
den3=[1, 2, 2]
sys3= control.TransferFunction(num3,den3)
```

##Plotando o gráfico

```
num_step = [1]
den_step = [1,0]
step = signal.TransferFunction(num_step,den_step)

t = np.linspace(0,4,100)

t,y1 = signal.step(signal.TransferFunction(num1,den1),0,t)
t,y2 = signal.step(signal.TransferFunction(num2,den2),0,t)
t,y3 = signal.step(signal.TransferFunction(num3,den3),0,t)

t,y_stp = signal.impulse(step,0,t) 

plt.plot(t,y_stp, "--y", label='degrau')
plt.plot(t, y1,  "-r", label='sistma 1')
plt.plot(t, y2,  "-k", label='sistema 2')
plt.plot(t, y3,  "-c", label='sistema 3')

plt. legend(loc = 'best')
plt.grid(axis='x', color='0.95')

plt.plot(t.take(y1.argmax()),y1.take(y1.argmax()), "-k", marker="o",label='Ponto máximo')
plt.plot(t.take(y2.argmax()),y2.take(y2.argmax()), "-k", marker="o",label='Ponto máximo')
plt.plot(t.take(y3.argmax()),y3.take(y3.argmax()), "-k", marker="o",label='Ponto máximo')
print("Cmax1 = ", y1.max(), "Tp = ", t[y1.argmax()])
print("Cmax2 = ", y2.max(), "Tp = ", t[y2.argmax()])
print("Cmax3 = ", y3.max(), "Tp = ", t[y3.argmax()])

plt.show()
```

##Analisando os sistemas

${G(s)=\dfrac{\omega_n^2}{s^2 + 2\zeta \omega_n + \omega_n^2}}$

${T_s = \dfrac{4}{\zeta \omega_n}}$

${T_p = \dfrac{\pi}{ \omega_d}}$

${\omega_n = \dfrac{\omega_d}{\sqrt{ 1 - \zeta^2} }}$

${\zeta = \dfrac{-ln(\%OS/100)} {\sqrt{\pi^2 + ln^2(\%OS/100)}}}$

${\%OS = \dfrac{c_{máx} - c_{final}}{c_{final}}}$

## Observando os três sistemas temos que ${\zeta \omega_n = 1}$ e portanto ${T_s = \dfrac{4}{1}= 4s}$ em ambos os casos.




```
#calculando wd
wd1=np.pi/t[y1.argmax()]
wd2=np.pi/t[y2.argmax()]
wd3=np.pi/t[y3.argmax()]

#calculando %OS
os1 = (y1.max()-1)*100
os2 = (y2.max()-1)*100
os3 = (y3.max()-1)*100

#Calculando zeta
z1 = - np.log(os1/100)/np.sqrt(np.pi**2 + np.log(os1/100)**2)
z2 = - np.log(os2/100)/np.sqrt(np.pi**2 + np.log(os2/100)**2)
z3 = - np.log(os3/100)/np.sqrt(np.pi**2 + np.log(os3/100)**2)

#Calculando wn
wn1= wd1/(np.sqrt(1-z1**2))
wn2= wd2/(np.sqrt(1-z2**2))
wn3= wd3/(np.sqrt(1-z3**2))

#Calculando Ts
ts1= 4/(z1*wn1)
ts2= 4/(z2*wn2)
ts3= 4/(z3*wn3)

#Printando os valores
print("Ts1 = ", ts1)
print("Ts2 = ", ts2)
print("Ts3 = ", ts3)
```

Os valores obtidos experimentalmente não são exatamente iguais ao teóricos devido ao método de obtenção das medidas que acarreta imprecisões.
